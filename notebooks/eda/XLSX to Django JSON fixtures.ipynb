````python
import json
import pandas as pd
import codecs
from matplotlib import pyplot as plt
````

````python
df = pd.read_excel("wine_web_data_style_processed.xlsx")
df_axis = pd.read_excel("wine_web_dr_encoded_wip.xlsx")

df[:2]
````

````python
def str_to_list(str):
    if not pd.isna(str):
        return eval(str)
    return []

def str_to_datefmt(str):
    # 28-04-2020
    if not pd.isna(str) and len(str.split('-')) == 3:
        d, m, y = str.split('-') 
        return f"{y}-{m}-{d}"
    return None

df_fix = df.copy()

df_fix['variedades'] = df['variedades'].apply(str_to_list)
df_fix['estilo'] = df['new_estilo2'].apply(str_to_list)
df_fix['boca'] = df['boca'].apply(str_to_list)
df_fix['color'] = df['color'].apply(str_to_list)
df_fix['aroma'] = df['aroma'].apply(str_to_list)

df_fix['fecha de cata'] = df['fecha de cata'].apply(str_to_datefmt)

df_fix['tipo_simplified'] = df_fix.tipo_simplified.str.title()

# df = df_fix
# del df_fix

df_fix.head(3)
````

````python
df_axis_eda = df_fix.groupby('id').first().join(df_axis.groupby('id').first())


# plt.figure(figsize=[12,8])
# df_axis_eda.groupby('zona de producción').mean()[['Dim1','Dim2','Dim3','Dim4','Dim5']].plot.bar(figsize=[16,8])
# df_axis_eda.groupby('tipo_simplified').mean()[['Dim1','Dim2','Dim3','Dim4','Dim5']].plot.bar(figsize=[16,8])

df_axis_eda.to_excel('wine_data_plus_dr.xlsx')
````

### Normalize Axis Values to range up to 1

````python
df_axis = df_axis.groupby('id').first()
df_axis = (df_axis / df_axis.max(axis=0)).round(3)
df_axis.head(3)
````

````python
df_axis['axii'] = df_axis[df_axis.columns[:6]].apply(lambda row: ','.join(row.values.astype(str)), axis=1)
df_axis[:5]
````

````python
# Append wine axis DR
df1 = df_fix.groupby('id').first()
df2 = df_axis[['axii']]

df_final = df1.join(df2).reset_index()
df_final.head(3)
````

````python
wine_data = []

for idx, row in df_final.iterrows():
    
    wine_data += [{
        "model": "api.wine",
        "pk": row.id,
        "fields": {
            "name": row['name'],
            "cellar_name": row.bodega,
            "year": row.year,
            "country": row.país,
            "zone": row['zona de producción'],
            "varieties": ','.join(row.variedades),
            "wine_type": row.tipo,
            "wine_type_simple": row.tipo_simplified,
            "price": row.price,
            "ecologic": row.ecologic,
            "penin_points": row.points,
            "tasting_date": row["fecha de cata"],
            "style": ','.join(row.estilo),
            "mouth": ','.join(row.boca),
            "color": ','.join(row.color),
            "smell": ','.join(row.aroma),
            "axii": row.axii
        }
    }]
    
    

fp = codecs.open('wine_data_web.json', 'w', 'utf8')
fp.write(json.dumps(wine_data).replace('NaN', 'null'))
fp.close()

wine_data[0]
````
